```python
import pandas as pd
import numpy as np

# Load the dataframes
df = pd.read_csv('/content/daily-shelter-overnight-service-occupancy-capacity-2024.csv')

# Filter the dataframe for Toronto, Men and Women sectors, and include LOCATION_ADDRESS and SECTOR
filtered_df = df[(df['LOCATION_CITY'] == 'Toronto') &
                 (df['SECTOR'].isin(['Men', 'Women']))].copy() # Use .copy() to avoid SettingWithCopyWarning

# Combine OCCUPIED_BEDS and OCCUPIED_ROOMS
filtered_df['occupancy_combined'] = filtered_df['OCCUPIED_BEDS'].fillna(filtered_df['OCCUPIED_ROOMS'])

# Combine CAPACITY_ACTUAL_BED and CAPACITY_ACTUAL_ROOM
filtered_df['capacity_combined'] = filtered_df['CAPACITY_ACTUAL_BED'].fillna(filtered_df['CAPACITY_ACTUAL_ROOM'])


# Calculate occupancy rate for combined columns
filtered_df['occupancy_rate_combined'] = (filtered_df['occupancy_combined'] / filtered_df['capacity_combined']) * 100

# Group by LOCATION_NAME and SECTOR and aggregate
neighbourhood_activity = filtered_df.groupby(['LOCATION_NAME', 'SECTOR'], as_index=False).agg({
    'occupancy_combined': 'sum',
    'capacity_combined': 'sum',
    'occupancy_rate_combined': 'mean',
    'LOCATION_ADDRESS': 'first' # Include location address in the aggregation
})

# Order by occupancy_combined from largest to smallest
neighbourhood_activity_sorted = neighbourhood_activity.sort_values(by='occupancy_combined', ascending=False)

display(neighbourhood_activity_sorted)
```

```python
from geopy.geocoders import Nominatim
import time

# Get unique location names
unique_locations = neighbourhood_activity_sorted['LOCATION_NAME'].unique()

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="shelter_geocoder")

# Create a list to store location and coordinates
location_coords = []

# Iterate through unique locations and get coordinates
for location in unique_locations:
    try:
        # Add "Toronto, Ontario" to the location name for better accuracy
        location_with_city = f"{location}, Toronto, Ontario"
        geocode_result = geolocator.geocode(location_with_city)
        if geocode_result:
            location_coords.append({
                'LOCATION_NAME': location,
                'latitude': geocode_result.latitude,
                'longitude': geocode_result.longitude
            })
        else:
            print(f"Could not geocode: {location}")
            location_coords.append({
                'LOCATION_NAME': location,
                'latitude': None,
                'longitude': None
            })
    except Exception as e:
        print(f"Error geocoding {location}: {e}")
        location_coords.append({
            'LOCATION_NAME': location,
            'latitude': None,
            'longitude': None
        })
    # Add a small delay to avoid hitting usage limits
    time.sleep(1)

# Create a DataFrame from the list
location_coords_df = pd.DataFrame(location_coords)

# Display the resulting DataFrame
display(location_coords_df)
```

## Calculate Distance to Weighted Centroid

```python
from geopy.distance import geodesic

# Define the coordinates of the weighted centroid
centroid_coords = (weighted_latitude, weighted_longitude)

# Calculate the distance from the centroid to each shelter
neighbourhood_activity_with_coords_cleaned['distance_to_centroid'] = neighbourhood_activity_with_coords_cleaned.apply(
    lambda row: geodesic((row['latitude'], row['longitude']), centroid_coords).km,
    axis=1
)

# Display the dataframe with the new distance column, showing LOCATION_ADDRESS and capacity_combined
display(neighbourhood_activity_with_coords_cleaned[['LOCATION_ADDRESS', 'capacity_combined', 'distance_to_centroid']].sort_values(by='distance_to_centroid'))

dis_to_centsss = neighbourhood_activity_with_coords_cleaned[['LOCATION_ADDRESS', 'capacity_combined', 'distance_to_centroid']].sort_values(by='distance_to_centroid')
```

```python
# Calculate the maximum daily capacity for each location
maximum_daily_capacity = filtered_df.groupby('LOCATION_ADDRESS')['capacity_combined'].max().reset_index()

# Merge the maximum daily capacity with the distances to centroid
dis_to_centsss_with_max_cap = pd.merge(dis_to_centsss, maximum_daily_capacity, on='LOCATION_ADDRESS', how='left')

# Rename the maximum capacity column for clarity
dis_to_centsss_with_max_cap.rename(columns={'capacity_combined_y': 'maximum_daily_capacity'}, inplace=True)

# Display the dataframe with location address, maximum daily capacity, and distance to centroid
display(dis_to_centsss_with_max_cap[['LOCATION_ADDRESS', 'maximum_daily_capacity', 'distance_to_centroid']].sort_values(by='distance_to_centroid'))
```

## Calculate Weighted Centroid

```python
# Calculate the weighted average of latitude and longitude
weighted_latitude = (neighbourhood_activity_with_coords_cleaned['latitude'] * neighbourhood_activity_with_coords_cleaned['occupancy_combined']).sum() / neighbourhood_activity_with_coords_cleaned['occupancy_combined'].sum()
weighted_longitude = (neighbourhood_activity_with_coords_cleaned['longitude'] * neighbourhood_activity_with_coords_cleaned['occupancy_combined']).sum() / neighbourhood_activity_with_coords_cleaned['occupancy_combined'].sum()

print(f"Approximate Center of Occupied Capacity (Weighted Centroid):")
print(f"Latitude: {weighted_latitude}")
print(f"Longitude: {weighted_longitude}")
```
